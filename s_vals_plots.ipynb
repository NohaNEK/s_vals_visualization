```python
import numpy as np
import matplotlib.pyplot as plt
id=5

# Generate synthetic data with a normal distribution
#data = np.random.normal(size=(500, 30))
data= s_all_gta[:,0,id].detach().cpu().numpy()
# Flatten the data to 1D for histogram plotting
flattened_data = data.flatten()
data2= s_all_cs[:,0,id].detach().cpu().numpy()
# Flatten the data to 1D for histogram plotting
flattened_data2 = data2.flatten()

# Plot the histogram
plt.hist(flattened_data, bins=30, density=True, alpha=0.7, color='red', edgecolor='gold',label='gta dist')
plt.hist(flattened_data2, bins=30, density=True, alpha=0.7, color='darkred', edgecolor='black',label='cs dist')

# Add a normal distribution fit to the histogram
mu, sigma = flattened_data.mean(), flattened_data.std()
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = np.exp(-(x - mu)**2 / (2 * sigma**2)) / (sigma * np.sqrt(2 * np.pi))
plt.plot(x, p, 'k', linewidth=2, color='gold')


mu2, sigma2 = flattened_data2.mean(), flattened_data2.std()
x = np.linspace(xmin, xmax, 100)
p = np.exp(-(x - mu2)**2 / (2 * sigma2**2)) / (sigma2 * np.sqrt(2 * np.pi))
plt.plot(x, p, 'k', linewidth=2)

plt.legend()

plt.title("Histogram of Normal Distribution of s"+str(id) +" of gta & cs(red channel)")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.show()
```
